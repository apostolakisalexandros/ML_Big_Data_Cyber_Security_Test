```python
import pandas as pd
import modin.pandas as pd
import pycaret
from pycaret.classification import *
```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/kdemertzis/ML_BigData_CyberSecurity-test/master/Project_Python/data_med.csv')
```

```python
grid = setup(data=dataset, target=dataset.columns[-1], html=False, silent=True, verbose=False, remove_outliers = True, normalize = True, normalize_method= 'minmax', multicollinearity_threshold = 0.8 )
```
